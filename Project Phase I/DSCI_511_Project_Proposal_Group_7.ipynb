### 3. Background on the Data Source
The Launch Library 2 API is maintained by the [Space Devs](https://thespacedevs.com/), an online community of developers who share an interest in and passion for spaceflight.<sup>2</sup> They [aim](https://thespacedevs.com/about) to maintain a free collection of REST APIs (following the guidelines for [representational state transfer-based web APIs](https://www.redhat.com/en/topics/api/what-is-a-rest-api)<sup>5</sup>) that present a broad array of data about multiple aspects of spaceflight, including rocket launches and relevant news articles.

For the information on launches, which we plan to focus on specifically, the Space Devs track all launches into orbit and numerous [suborbital launches](https://github.com/TheSpaceDevs/Tutorials/blob/main/faqs/faq_LL2.md#which-suborbital-launches-are-allowed-in-ll2) that reach a sufficiently high altitude and are live-streamed on the internet, with occasional others based on popular interest.<sup>6</sup> Should we want to expand the reach of our project beyond just rocket launches, the Space Devs also offer APIs that focus on space events (e.g., spacecraft landings or spacewalks), space agencies, launch vehicles, and more.

Our data was created from reports of spacecraft launches and associated communications. According to the [Frequently Asked Questions](https://github.com/TheSpaceDevs/Tutorials/blob/main/faqs/faq_LL2.md#where-does-ll2-get-its-data-from) page, the data in Launch Library 2 stems from three main types of public data sources:<sup>6</sup>
1. First-party data: Direct from space agencies, launch service providers, satellite operators, etc.
2. Second-party data: From news outlets (that are known to be reliable)
3. Community data: From forums and user reports, which are the least reliable of these three data sources.

Some historic data is sourced from other public databases, as well. The data provided by the API is updated on a scale of several times a day to several times an hour, depending on the incidence of launches.<sup>6</sup>

LaunchLibrary API v2.3.0 is the latest supported version of the API. It is [free to use](https://ll.thespacedevs.com/docs/#/launches), though subject to rate limiting, with a maximum of 15 calls per hour per IP address for requests without an authentication token.<sup>1</sup> If desired, users can subscribe to the Space Devs’ [Patreon](https://www.patreon.com/TheSpaceDevs) or send a one-time monetary donation to obtain an API key with a higher rate limit. In addition, for the project's development phase, they offer a specific URL for development testing (https://lldev.thespacedevs.com) with limited, older data but no rate limits.

#### 4.2 Data Acquisition via API Phase
In the Data Acquisition Phase, the initial task is to write the full Python script using the Requests library to create the GET requests to the API endpoint. This would also include running a sample of the code for the first 5-10 entries to verify the API request's functionality before executing the full download. An important aspect of the GET requests is handling the pagination process in accordance with the Launch Library 2 API documentation.<sup>7</sup> Which we plan to accomplish with the use of a while loop to call the next URL provided in each JSON response. Another important consideration of the script is that it must operate within the rate limit of 15 calls per hour. According to our math, the API has a minimum count of 7,657 previous launches as illustrated in the [sample JSON Response](https://ll.thespacedevs.com/2.3.0/launches/?format=json), with a maximum of 100 per call (7657 launches / 100 launches per call), resulting in 77 API calls to retrieve all the data.<sup>7</sup> Since we are only allowed 15 calls per hour, we need to pause the code for one hour after every 14 calls. Therefore, we plan to use the Time module's sleep command to accomplish this. It is estimated that retrieving all the data will require nearly 6 hours (77 API calls / 14 calls per hour).

Another key task of this Data Acquisition phase is to save and store the raw data. Once the total number of historical launches we want to include is retrieved, the complete, untouched raw data will be saved as a single CSV file. This will serve as our stable baseline, from which each team member can utilize their individual data cleaning scripts to clean a portion of the data in the next phase. Standardizing the process of running the same data-cleaning scripts from the same baseline data will always produce the identical final dataset, ensuring reproducibility of the project. The baseline raw data will remain unaltered and stored in the group’s shared repository. This “always-keep-one-copy-of-the-raw-data” approach reduces the risk of corruption to the baseline and minimizes the number of API access calls.

### 5. Data Sample
We were able to perform a preliminary test of the process for making a GET request to the Launch Library 2 API and retrieving a raw data sample.


First steps:
* Import requests library
* Define the API endpoint URL
* Make the GET request

```python
import requests # import library to send HTTP requests

# The specific URL for 'previous launches' endpoint from the LL2 documentation
# Add '?limit=5' to the end to ask for only 5 results for our sample
url = "https://ll.thespacedevs.com/2.3.0/launches/previous/?limit=5"

# Send the GET request to API
response = requests.get(url)
```

Next, check if the request was successful:

```python
# A status code of 200 means the request was successful
print(response.status_code)
```

Finally, get the raw data and take a look at what we received:

```python
from pprint import pprint # import library for allow easy viewing

# Parse the JSON text from the response into a Python dictionary
raw_data = response.json()

# The launch records are inside the 'results' key
launch_sample = raw_data['results']

# Print the sample to the screen
pprint(launch_sample)
```

### 6. Key Data Transformations, Limitations, and Challenges
If incomplete or inconsistent data is identified during data processing, we can utilize various methods to enforce consistency and calculate reasonable values of variables. For instance, in the case of the launch’s payload, estimating mass would be a difficult task if this information is not complete from our data sources. We could put in place a transformation rule-based system that employs the following techniques when precise payload mass is not given (which is typical in commercial manifests):

* Estimates of satellite bus mass are derived from known bus kinds and public satellite dimensions.
* Use standard mass values for standardized small-satellite classes (such as CubeSats).

There are several additional ways in which we may wish to standardize our data:
* **Temporal Standardization:** Standardizing the time and date to the international standard, ISO 8601, UTC, is our intended goal.
 * **Entity Resolution:** Rockets' and organizations' names will be clarified; for example, "SpaceX," "Space Exploration Technologies," and "SpaceX (USA)" will all resolve to the same organization identifier.
* **Conflict resolution and data fusion:** If we ultimately scrape any data from the Next Spaceflight website, we will create a trust hierarchy for our sources. For instance, if there are payload mass disputes between Launch Library 2 and Next Spaceflight, the value from the more reliable source, likely the API, will be selected. Then a flag will indicate that the conflict has been resolved.

#### 6.1 Further Limitations to the Dataset
Fortunately, our primary data source contains a wide breadth of information related to rocket launches. As mentioned previously, if we do find that some key aspects are missing/incomplete in the Launch Library 2 data, we may attempt web scraping from the Next Spaceflight website<sup>3</sup> to obtain supplementary information; however, this would bring its own challenges in terms of combining the data from the two sources, which may differ in formatting and terminology.
