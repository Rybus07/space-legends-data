**Purpose:** This notebook contains the Python code to make GET requests to the Launch Library 2 API, retrieve a raw data sample, and eventually acquire the full dataset.


# 1st Steps:
Import requests Library

Define the API Endpoint URL

Make the GET request

```python
import requests # import library to send HTTP requests

# The specific URL for 'previous launches' endpoint from the LL2 documentation
# Add '?limit=5' to the end to ask for only 5 results for our sample
url = "https://ll.thespacedevs.com/2.3.0/launches/previous/?limit=5"

# Send the GET request to API
response = requests.get(url)
```

# Next:
Check if the request was successful

```python
# A status code of 200 means the request was successful
print(response.status_code)
```

# Finally:

Get the raw data

and take a look at what we received

```python
from pprint import pprint # import library for allow easy viewing

# Parse the JSON text from the response into a Python dictionary
raw_data = response.json()

# The launch records are inside the 'results' key
launch_sample = raw_data['results']

# Print the sample to the screen
pprint(launch_sample)
```
