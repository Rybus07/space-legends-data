## Test Data Loop

With the few parameters that we have acqired lets try to see if we can grab the specific information that we want and be able to create it into a format that is easily converted into a CSV and/or Pandas dataframe
* We should decide if we just want to do one big CSV/dataframe, or several different ones focusing on different topics that are connected by e.g. some sort of tagID

Here, using the data_sample of just 100 launches for expediency, I've written a loop to create a table with launch-related parameters, examined the table as a dataframe, and saved the table as a TSV

```python
#Referencing lecture notes 4.1.2
launch_table = [] #intialize - just doing a table of launch-related parameters from cell above for now

#TODO: These headers and the data gathered in loop subject to change based on what we decide we want!
header = ["Date_of_Launch", "Lauch_Status", "Short_Form_Status", "Launch_pad_location_name", "Launch_pad_name", "Launch_pad_country_name",
   "Launch_pad_country_ID", "Launch_pad_lattitude", "Launch_pad_longitude"]
#launch_table.append(header)

#TODO: we could also do some preliminary cleaning in this loop if we want
for launch in data_sample:
    Date = launch['net']
    Status = launch['status']['name']
    SF_status = launch['status']['abbrev']
    Pad_location = launch['pad']['location']['name']
    Pad_name = launch['pad']['name']
    Pad_country_name = launch['pad']['country']['name']
    Pad_country_ID = launch['pad']['country']['id']
    Latt = launch['pad']['latitude']
    Long = launch['pad']['longitude']

    row = [Date, Status, SF_status, Pad_location, Pad_name, Pad_country_name, Pad_country_ID, Latt, Long]
    launch_table.append(row)
```

```python
#Referenced Stack Overflow on how to turn list of lists into Pandas dataframe: 
#https://stackoverflow.com/questions/32857544/nested-list-to-pandas-dataframe-with-headers

import pandas as pd
df = pd.DataFrame(launch_table,columns=header)

df.head()
```

```python
#Since launch_pad_location_name has commas, not ideal to save as a CSV - saving it as a TSV instead
#TODO: add a header?
with open("data_test/launch_table_test.tsv", "w") as f: 
    for row in launch_table:

        #Everything in the TSV needs to be a string, so do that conversion here:
        row_string = []
        for element in row: 
            if type(element) != str: 
                element_string = str(element)
                row_string.append(element_string)
            else:
                row_string.append(element) 

        #Now actually write to the tsv:
        f.write("\t".join(row_string) + "\n") #joining elements together w/ a tab
```
