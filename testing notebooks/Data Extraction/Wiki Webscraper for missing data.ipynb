```python
import requests
from bs4 import BeautifulSoup

#With looking into the Wikipedia Rest API Documentation it appears that when requesting based on title it requires the title separated with "%20" rather than a space
def get_parsed_page(page_title):
    '''
    The fucntion allows us to use a page title name and be able to obtain the parsed information of the page.
    '''
    #Wikipedia requires a User-Agent to use their rest API, so I followed Lecture 3 Notes and this link from stack overflow on how to implement
    #https://stackoverflow.com/questions/79618466/specify-contact-info-in-wikipedia-user-agent-policy
    headers = {"User-Agent": "FireFOX/1.0 (rap369@drexel.edu)"}
    #URL Creator
    api_request_url = "https://en.wikipedia.org/api/rest_v1/page/html/"
    request_url = api_request_url + page_title
    
    #API request
    api_response = requests.get(request_url, headers=headers)
    api_text = api_response.text

    #BeautifulSoup Parser
    wiki_soup = BeautifulSoup(api_text, 'html.parser')

    return api_text, wiki_soup
```
