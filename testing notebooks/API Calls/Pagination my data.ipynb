```python
from datetime import datetime
import dateutil.parser as dateparser
from pprint import pprint
import requests
import json
import time
```

```python
def query_space_launches():
    # Set 5-year date range
    start_time = dateparser.parse("1/1/2015")
    end_time = dateparser.parse("12/31/2019")

    print("Start date:", start_time.isoformat())
    print("End date:", end_time.isoformat())
```

```python
  # Base parameters
params = {
        'net__gte': start_time.isoformat(),
        'net__lte': end_time.isoformat(),
        'mode': 'detailed',
        'limit': 100,
        'ordering': 'net'}
```

```python
   all_results = []
    current_url = "https://ll.thespacedevs.com/2.3.0/launches/previous/"
    page = 1
    api_calls_made = 0
    total_time_taken = 0
    time_between_calls = []
```

```python
 # Track overall script execution time
    script_start_time = time.time()
    
    print("Querying API for 5 years of launch data...")
    
    while current_url:
        print(f"Fetching page {page}...")
```

```python
        # Track time before API call
        call_start_time = time.time()
        
        try:
            if page == 1:
                response = requests.get(current_url, params=params, timeout=60)
            else:
                response = requests.get(current_url, timeout=60)
            
            # Track API call completion time
            call_end_time = time.time()
            call_duration = call_end_time - call_start_time
            total_time_taken += call_duration
            api_calls_made += 1
            print(f"Page {page}: Retrieved {len(response.json().get('results', []))} launches in {call_duration:.2f} seconds")
```

```python
if response.status_code == 200:
                data = response.json()
                results = data.get('results', [])
                all_results.extend(results)
                
                # Check for next page
                current_url = data.get('next')
                page += 1
                
                # Respect rate limits - wait between pages if needed
                if current_url:
                    wait_time = 1  # seconds between calls
                    print(f"Waiting {wait_time} second before next page...")
                    time_between_calls.append(wait_time)
                    time.sleep(wait_time)
                else:
                    print(f"Error on page {page}: Status code {response.status_code}")
                break
except Exception as e:
       call_end_time = time.time()
       call_duration = call_end_time - call_start_time
       total_time_taken += call_duration
        api_calls_made += 1
            
        print(f"Error on page {page}: {e}")
         break
 
```

```python
  # Calculate total script execution time
    script_end_time = time.time()
    total_script_time = script_end_time - script_start_time
    
    # Calculate average times
    avg_time_per_call = total_time_taken / api_calls_made if api_calls_made > 0 else 0
    avg_time_between_calls = sum(time_between_calls) / len(time_between_calls) if time_between_calls else 0
```

```python
# Create final data structure with comprehensive metadata
    final_data = {
        'metadata': {
            'query_parameters': {
                'start_date': start_time.isoformat(),
                'end_date': end_time.isoformat(),
                'date_range_years': 5,
                'mode': 'detailed',
                'results_per_page': 100
            },
            'execution_metrics': {
                'total_api_calls_made': api_calls_made,
                'pages_retrieved': page - 1,
                'total_launches_retrieved': len(all_results),
                'timestamp_completed': datetime.now().isoformat(),
                'script_execution_time_seconds': round(total_script_time, 2),
                'total_api_call_time_seconds': round(total_time_taken, 2),
                'average_time_per_api_call_seconds': round(avg_time_per_call, 2),
                'time_between_api_calls_seconds': time_between_calls,
                'average_time_between_calls_seconds': round(avg_time_between_calls, 2),
                'rate_limit_respected': True if time_between_calls else False
            },
            'performance_metrics': {
                'calls_per_hour': api_calls_made / (total_script_time / 3600) if total_script_time > 0 else 0,
                'launches_per_second': len(all_results) / total_script_time if total_script_time > 0 else 0,
                'data_retrieval_efficiency': round(total_time_taken / total_script_time * 100, 2) if total_script_time > 0 else 0
            }
        },
        'results': all_results
    }
```

```python
  # Save to JSON file
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"space_launches_5years_with_metrics_{timestamp}.json"
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(final_data, f, indent=2, ensure_ascii=False)
```

```python
   # Print summary
    print("\n" + "="*60)
    print("QUERY EXECUTION SUMMARY")
    print("="*60)
    print(f"Total API calls made: {api_calls_made}")
    print(f"Pages retrieved: {page - 1}")
    print(f"Total launches retrieved: {len(all_results)}")
    print(f"Total script execution time: {total_script_time:.2f} seconds")
    print(f"Total API call time: {total_time_taken:.2f} seconds")
    print(f"Average time per API call: {avg_time_per_call:.2f} seconds")
    print(f"Average time between calls: {avg_time_between_calls:.2f} seconds")
    print(f"Data saved to: {filename}")
    print("="*60)
    
    return final_data
```
